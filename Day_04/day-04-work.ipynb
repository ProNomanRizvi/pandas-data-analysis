# **Adding a column**

```python
import pandas as pd
```

### Preview Data\nViewing first/last rows of the dataset

```python
data = {
    'Name': [
        'Alice', 'Bob', 'Charlie', 'David', 'Eve',
        'Frank', 'Grace', 'Hank', 'Ivy', 'Jack',
        'Kathy', 'Leo', 'Mia', 'Noah', 'Olivia',
        'Paul', 'Quinn', 'Rose', 'Sam', 'Tina'
    ],
    'Age': [
        25, 30, 35, 28, 22, 40, 31, 29, 24, 33,
        27, 38, 26, 32, 23, 45, 29, 34, 21, 36
    ],
    'City': [
        'New York', 'San Francisco', 'Los Angeles', 'Chicago', 'Miami',
        'Seattle', 'Austin', 'Boston', 'Denver', 'Atlanta',
        'Portland', 'Dallas', 'Phoenix', 'San Diego', 'Philadelphia',
        'Las Vegas', 'Houston', 'Detroit', 'Minneapolis', 'Nashville'
    ],
    'Salary': [
        50000, 60000, 70000, 55000, 48000,
        85000, 62000, 58000, 52000, 67000,
        54000, 78000, 51000, 64000, 49000,
        95000, 59000, 72000, 47000, 81000
    ]
}

df = pd.DataFrame(data)
print(df.head())
```

## **01- Using Square Brackects**

```python
# Add Bounus Column
df["Bonus"] = df["Salary"] * 0.1
print(df.head())
```

### Preview Data\nViewing first/last rows of the dataset

```python
# Add Total Salary (Salary + Bounus)
df["TotalSalary"] = df["Salary"] + df["Bonus"]
print(df.head())
```

## **02- Using **`insert()`** method**

```python
sample_data = {
    'Name': ['John', 'Jane', 'Bob', 'Alice'],
    'Age': [25, 30, 35, 40],
    'City': ['New York', 'San Francisco', 'Los Angeles', 'Chicago']
}
dataFrame = pd.DataFrame(sample_data)
print(dataFrame)
```

### Insert Column\nAdding new column at specific position

```python
# I want to add ID Column at first
emp_id = [101, 102, 103, 104]
dataFrame.insert(0, 'EmpID', emp_id)
print(dataFrame)
```

### Insert Column\nAdding new column at specific position

```python
# Add Country
country = ['USA', 'USA', 'USA', 'USA']
dataFrame.insert(4, 'Country', country)
print(dataFrame)
```

# **Update the Data**
using **`loc()`**

```python
sample_data = {
    'Name': [
        'Alice', 'Bob', 'Charlie', 'David', 'Eve',
        'Frank', 'Grace', 'Hank', 'Ivy', 'Jack',
        'Kathy', 'Leo', 'Mia', 'Noah', 'Olivia',
        'Paul', 'Quinn', 'Rose', 'Sam', 'Tina'
    ],
    'Age': [
        25, 30, 35, 28, 22, 40, 31, 29, 24, 33,
        27, 38, 26, 32, 23, 45, 29, 34, 21, 36
    ],
    'City': [
        'New York', 'San Francisco', 'Los Angeles', 'Chicago', 'Miami',
        'Seattle', 'Austin', 'Boston', 'Denver', 'Atlanta',
        'Portland', 'Dallas', 'Phoenix', 'San Diego', 'Philadelphia',
        'Las Vegas', 'Houston', 'Detroit', 'Minneapolis', 'Nashville'
    ],
    'Salary': [
        50000, 60000, 70000, 55000, 48000,
        85000, 62000, 58000, 52000, 67000,
        54000, 78000, 51000, 64000, 49000,
        95000, 59000, 72000, 47000, 81000
    ]
}
DF = pd.DataFrame(sample_data)
print(DF.head())
```

### Preview Data\nViewing first/last rows of the dataset
