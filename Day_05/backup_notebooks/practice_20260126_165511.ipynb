```python
import pandas as pd
```

```python
data = "/content/drive/MyDrive/Pandas for Data Analysis/Day_05/practice_data.csv"
df = pd.read_csv(data)
print(df.head())
```

```python
print(df.info())
```

```python
print(df.isnull().sum())
```

```python
print(df['Transaction_ID'])
```

Fill Transaction_ID Column

```python
df['Transaction_ID'].fillna(1004, inplace=True)
print(df.head())
```

Fill Date Column

```python
df['Date'].fillna(df['Date'].mode()[0], inplace=True)
print(df)
```

Fill Customer_Name Column

```python
df['Customer_Name'].fillna(df['Customer_Name'].mode()[0], inplace=True)
print(df)
```

Fill Product_Catagory Column

```python
df['Product_Category'].fillna(df['Product_Category'].mode()[0], inplace=True)
print(df)
```

Fill Amoount_Spent Column

```python
df['Amount_Spent'].fillna(df['Amount_Spent'].mean(), inplace=True)
print(df)
```

Fill Rating Column

```python
# Replace Unknown with Nan
df['Rating'].replace('Unknown', pd.NA, inplace=True)
```

```python
# Convert Rating Column to Numeric
df['Rating'] = pd.to_numeric(df['Rating'])
```

```python
print(df['Rating'])
```

```python
# Fill Values
df['Rating'].fillna(df['Rating'].mean(), inplace=True)
print(df)
```

Fill Member Column

```python
df['Member'] = df['Member'].replace("Unknown", pd.NA)
df['Member'].fillna(df['Member'].mode()[0], inplace=True)
```

```python
print(df['Member'])
```
