# **Sorting**

```python
import pandas as pd
```

Load Data

```python
data = "unsorted_data.csv"
df = pd.read_csv(data)
print(df.head())
```

### **Sort by one column**

```python
# Sort the DataFrame by 'EmployeeID' in ascending order
df.sort_values(by='EmployeeID', inplace=True, ascending=True)
print(df.head())
```

```python
# Sort the DataFrame by FirstName in ascending order
df.sort_values(by='FirstName', inplace=True, ascending=True)
print(df.head())
```

### **Sort Multiple Columns**

```python
print(df.head())
```

```python
# Sort the DataFrame by "EmployeeID", "Salary", and "JoinDate" in ascending order
df.sort_values(by=['EmployeeID', 'Salary', 'JoinDate'], inplace=True, ascending=[True, True, True])
print(df.head())
```

# **Aggregation**

```python
# Apply Summry Statistics on the DataFrame
summary = df.describe()
print(summary)
```

```python
# Operations on Salary Column
salaries = df['Salary']
print("Salaries Column:")
print(salaries)
```

```python
print("Sum of Salaries:", salaries.sum())
print("Mean of Salaries:", salaries.mean())
print("Median of Salaries:", salaries.median())
print("Standard Deviation of Salaries:", salaries.std())
print("Minimum Salary:", salaries.min())
print("Maximum Salary:", salaries.max())
```
