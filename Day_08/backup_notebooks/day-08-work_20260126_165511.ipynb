# **Grouping**

```python
import pandas as pd
```

Load Data

```python
data_path = "students_data.xlsx"
df_students = pd.read_excel(data_path)
print(df_students.head())
```

### **01- By One Column**

```python
grouped_by_Attendance = df_students.groupby('Grade')['GPA'].max()
print(grouped_by_Attendance)
```

### **02- By Multi Columns**

```python
print(df_students.head())
```

```python
# Group by Grade and Attendance, then calculate mean GPA
grouped = df_students.groupby(['Grade', 'Attendance %'])['GPA'].mean()
print(grouped)
```
