# **Group Data (Practice)**

```python
import pandas as pd
```

### Load Data
Reading data from file into DataFrame

```python
data_path = "sales_data.csv"
df = pd.read_csv(data_path)
df.head()
```

### Group Data
Aggregating data by groups

```python
# Group by "store_regin" and calculate total sales
grouped = df.groupby('store_region')['sales'].sum()
print(grouped)
```

### Group Data
Aggregating data by groups

```python
# Group by "Product" and "Sales", then calculate total profit
grouped_multi = df.groupby(['product', 'sales'])['profit'].sum()
print(grouped_multi)
```

### Group Data
Aggregating data by groups

```python
# Check profit by Date
profit_by_date = df.groupby('date')['profit'].sum()
print(profit_by_date)
```
