# Day 10: DataFrame Concatenation
## Combining multiple DataFrames using pd.concat()

### Import Libraries
Importing pandas for data manipulation

```python
import pandas as pd
```

### Create First Employee Dataset
Employee data with US cities (10 records)

```python
employee_data_1 = {
    "Name": ["John", "Jane", "Jack", "Jill", "Ali", "Iqbal", "Hassan", "Hina", "Hafsa", "Hassan"],
    "Age": [25, 30, 35, 40, 45, 50, 55, 60, 65, 70],
    "City": ["New York", "Los Angeles", "Chicago", "Houston", "Phoenix", "Philadelphia", "San Antonio", "San Diego", "Dallas", "San Jose"],
    "Salary": [50000, 60000, 70000, 80000, 90000, 100000, 110000, 120000, 130000, 140000]
}
df_employee_1 = pd.DataFrame(employee_data_1)
df_employee_1
```

### Create Second Employee Dataset
Employee data with Pakistani cities (5 records)

```python
employee_data_2 = {
    "Name": ["Noman", "Ali", "Ahmed", "Usman", "Hassan"],
    "Age": [25, 30, 35, 40, 45],
    "Salary": [50000, 60000, 70000, 80000, 90000],
    "City": ["Karachi", "Lahore", "Islamabad", "Peshawar", "Quetta"]
}
df_employee_2 = pd.DataFrame(employee_data_2)
df_employee_2
```

### Vertical Concatenation (axis=0)
Stacking DataFrames row-wise to combine all employee records

```python
df_concate_ver = pd.concat([df_employee_1, df_employee_2], axis=0, ignore_index=True)
df_concate_ver
```

### Horizontal Concatenation (axis=1)
Merging DataFrames column-wise, creating NaN values for mismatched rows

```python
df_concate_hor = pd.concat([df_employee_1, df_employee_2], axis=1, ignore_index=True)
df_concate_hor
```
