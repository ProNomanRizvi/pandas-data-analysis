```python
import pandas as pd
```

```python
# Create the dummy data for Day 13
data = {
    'Department': ['HR', 'HR', 'IT', 'IT', 'Sales', 'Sales'],
    'Gender': ['Female', 'Male', 'Female', 'Male', 'Female', 'Male'],
    'Salary': [60000, 62000, 80000, 85000, 70000, 72000],
    "Date": ["2022-01-01", "2022-01-02", "2022-01-03", "2022-01-04", "2022-01-05", "2022-01-06"],
    "Company": ["Company A", "Company B", "Company C", "Company D", "Company E", "Company F"],
    "Price": [100, 200, 300, 400, 500, 600]
}

df = pd.DataFrame(data)
```

```python
# Create a pivot table showing average Salary by Department and Gender
pivot_df = df.pivot_table(
    values='Salary', 
    index='Gender', 
    columns='Department', 
    aggfunc='mean'
)
pivot_df
```

```python
# Count how many employees of each Gender are in each Department
count_df = pd.crosstab(df['Department'], df['Gender'])
count_df
```

```python
# Show percentages instead of counts (normalize)
percent_df = pd.crosstab(df['Department'], df['Gender'], normalize='index')
percent_df
```

```python
data = {
    "Date": ["2026-01-01", "2026-01-02", "2026-01-03"],
    "Apple": [150, 152, 155],
    "Google": [2800, 2810, 2825],
    "Microsoft": [300, 305, 310]
}

# Create a DataFrame
df_stock_prices = pd.DataFrame(data)
df_stock_prices
```

```python
# Melt the dataframe
melted_df = df_stock_prices.melt(
    id_vars=['Date'],        # Column(s) to keep as identifier variables
    var_name='Company',      # Name for the new "variable" column
    value_name='Price'       # Name for the new "value" column
)
melted_df
```
