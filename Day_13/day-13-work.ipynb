```python
import pandas as pd
```

```python
df_weather = pd.read_csv("weather_data.csv")
df_weather.head()
```

```python
df_weather.pivot(index="date", columns="city")
```

```python
print("Only Humidity: \n")
df_weather.pivot(index="date", columns="city", values="humidity")
```

```python
print("Only Temperature: \n")
df_weather.pivot(index="date", columns="city", values="temperature")
```

```python
df = pd.read_csv("weather_data_extended.csv")
df.head()
```

```python
print("Humidity: \n")
df.pivot_table(index="city", columns="date", values="humidity")
```

```python
print("Temperature: \n")
df.pivot_table(index="city", columns="date", values="temperature")
```

```python
df.pivot_table(index="city", columns="date", aggfunc="sum")
```

```python
df.pivot_table(index="city", columns="date", aggfunc="mean")
```

```python
df.pivot_table(index="city", columns="date", aggfunc="count")
```

```python
df.pivot_table(index="city", columns="date", aggfunc="diff")
```

```python
df.pivot_table(index="city", columns="date", margins=True)
```

```python
df2 = pd.read_csv("combined_weather.csv")
df2.head()
```

```python
df2["date"] = pd.to_datetime(df["date"])
type(df2['date'][0])
```
